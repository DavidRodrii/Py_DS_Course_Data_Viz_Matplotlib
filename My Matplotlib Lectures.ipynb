## Object Oriented API Method

Instantiate figure objects then call methods or attributes from object.

So we create figure objects, then call methods to them

```python
fig = plt.figure()
# think of figure object as blank canvas
# We can add a set of Axes by passing it as a List
# List takes in 4 Arguments: Left, Bottom, Width, Height
# Ranging from 0 - 1
axes = fig.add_axes([0.1,0.1,0.8,0.8])
# Now that we have the Axes made, we can plot on them
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

Object Oriented Method gives us much more control over our parameters, but it does take a few more steps

### Subplots via Object Oriented Method

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

# Matplotlib Part 2

### Subplots w/ Object Oriented Method

```python
fig,axes = plt.subplots(nrows=3,ncols=3)
plt.tight_layout()
# axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

plt.tight_layout( ) takes care of any overlapping that may occur while displaying plots

## Figure Size, Aspect Ratio, and DPI

```python
fig = plt.figure(figsize=(3,2),dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

## Legends

Adding a Legend is a two step process

step 1.) ax.legend( )

step 2.) define labels inside ax.plot( ) commands

ax.plot(x, x**2, label='X Squared')

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')

ax.legend()
# Since the legend is overlapping the plotted line we can fix this
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')
# This is fixed by specifying the loc = 'Location Code'

# Location Codes can be found on Matplotlib Documentation online
# loc = 2 ---> 'Top Left'

ax.legend(loc=2)
```

```python
# If we need more precise control over the location of the legend
# We can pass in a tuple of x and y coordinates from the bottom left corner
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')

ax.legend(loc=(0.1,0.1))
```

# Matplotlib Part 3

covers colors, line widths, line types

## Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# We can edit the color of the plot via the color parameter
# we can pass in either color strings like 'red' or 'orange'
# But we can also pass in RGB Hex Codes as well such as
# '#FF8C00' for different custom colors
ax.plot(x,y,color='#FF8C00')
```

We can use google search to find some sort of RGB Hex Code maker tool online to use our own custom colors in our plots

### Line width/Line Style

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',linewidth=5,alpha=0.3)
```

linewidth = # is obvious, we can also type in lw = # for shorthand

If we want transparent lines, we can use the alpha = # argument to pass in a number between 0 and 1

linestyle = '--' is another argument that in this case indicates dashed lines, with ls = '--' for shorthand as well 

### Markers

If we want to specify on the plot where each point lies, we can use the argument of marker = 'o'

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=3,ls='-',marker='o')
```

```python
# We can also use markersize to change the size of the markers
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=3,ls='-',marker='o',markersize=10)
```

we can also specify the marker face color, marker edge width, and marker edge color

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=1,ls='-',marker='o',markersize=15,
       markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green')
```

### Control over Axis Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=2,ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
